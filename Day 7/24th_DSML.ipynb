```
import numpy as np
x = np.array([2.2,4.3,5.5,6.3,8])
y = np.array([3.1,4.5,7.6,5.4,8.9])
```

```
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()
```

```
m = 4.3
c = 2
yp = m*x + c
```

```
plt.scatter(x,y)
plt.plot(x,yp,c='r')
plt.show()
```

```
e = ((y-yp)**2).mean()
e
```

```
dem = 2*np.dot((y-yp).T,(-x))
dec = (-2*(y-yp)).sum()
```

```
lr = 1e-3
print(1e-4*dem , 1e-4*dec)
```

```
m = m - lr * dem
c = c - lr * dec
```

```
yp = m*x + c
e = ((y-yp)**2).mean()
e
```

```
plt.scatter(x,y)
plt.plot(x,yp,c='r')
plt.show()
```

```
lr = 1e-3
for i in range(35):
  yp = m*x + c

  dem = 2*np.dot((y-yp).T,(-x))
  dec = (-2*(y-yp)).sum()

  m = m - lr * dem
  c = c - lr * dec

  e = ((y-yp)**2).mean()
  print(i , e)
```

```
plt.scatter(x,y)
plt.plot(x,yp,c='r')
plt.show()
```

**For Multi variables**

```
X = df[['TV','radio']].values
Y = df[['sales']].values
```

```
m = np.array([[0.5],[0.7]])
c = 0.5
```

```
print(m[0][0])
print(m[1][0])
```

```
Yp = np.dot(X,m) + c
```

```
((Y - Yp)**2).mean()
```

```
dem1 = 2*(np.dot((Y-Yp).T,-X[:,0:1]))
dem2 = 2*(np.dot((Y-Yp).T,-X[:,1:2]))
dec = (-2*(Y-Yp)).sum()
```

```
lr = 1e-7
print(lr*dem1,lr*dem2,lr*dec)
```

```
m[0][0] = m[0][0] - lr * dem1
m[1][0] = m[1][0] - lr * dem2
c = c - lr * dec
```

```
Yp = np.dot(X,m) + c
((Y - Yp)**2).mean()
```

```
for i in range(100):
  Yp = np.dot(X,m) + c

  dem1 = 2*(np.dot((Y-Yp).T,-X[:,0:1]))
  dem2 = 2*(np.dot((Y-Yp).T,-X[:,1:2]))
  dec = (-2*(Y-Yp)).sum()

  m[0][0] = m[0][0] - lr * dem1
  m[1][0] = m[1][0] - lr * dem2
  c = c - lr * dec

  error = ((Y - Yp)**2).mean()
  print(i,error)
```

```
for i in range(200):
  Yp = np.dot(X,m) + c

  dem1 = 2*(np.dot((Y-Yp).T,-X[:,0:1]))
  dem2 = 2*(np.dot((Y-Yp).T,-X[:,1:2]))
  dec = (-2*(Y-Yp)).sum()

  m[0][0] = m[0][0] - lr * dem1
  m[1][0] = m[1][0] - lr * dem2
  c = c - lr * dec

  error = ((Y - Yp)**2).mean()
  print(i,error)
```
