```
import pandas as pd
df = pd.read_csv('diabetes.csv')
```

```
df.head(20)
```

```
df.groupby('Outcome')['SkinThickness','Insulin','BloodPressure'].mean()
```

```
df0 = df[df['Outcome'] == 0]        #Non-Diabetic
df1 = df[df['Outcome'] == 1]        #Diabetic
```

```
df0['SkinThickness'] = df0['SkinThickness'].replace(0,19.66)
df1['SkinThickness'] = df1['SkinThickness'].replace(0,22.16)
```

```
df0['Insulin'] = df0['Insulin'].replace(0,68.79)
df1['Insulin'] = df1['Insulin'].replace(0,100.34)
```

```
df0['BloodPressure'] = df0['BloodPressure'].replace(0,68.18)
df1['BloodPressure'] = df1['BloodPressure'].replace(0,70.82)
```

```
df = df0.append(df1).reset_index().drop(['index'],axis=1)
```

```
(df['BloodPressure'] == 0).sum()
```

```
df.head(15)
```

```
#Features
X = df.drop(['Outcome'],axis=1)
#Label
Y = df[['Outcome']]
```

```
#Training and testing split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y)
```

```
from sklearn.tree import DecisionTreeClassifier
dmodel = DecisionTreeClassifier()
```

```
dmodel.fit(xtrain,ytrain)
```

```
#Training accuracy
print(dmodel.score(xtrain,ytrain))
#Testing Accuracy
print(dmodel.score(xtest,ytest))
```

```
groups = [[['x1', 1], ['x2', 1]], [['x3', 0], ['x4', 0]]]

classes = [0, 1]

print(groups)
```

```
instances = 0

instances += sum([len(group) for group in groups])

print(instances)
```

```
#initilize
gini = 0.0

for group in groups:
    #Define no. of elements in each group
    size = float(len(group))
    #skip the instances in case if any group has been created 
    #without any value
    if (size == 0):
        continue
    
    #Calculate the gini score
    score = 0.0
    for class_val in classes:
        p = [row[-1] for row in group].count(class_val) / size
        score += p * p
        
    gini += (1.0 - score) * (size / instances)
print(gini)
```

```
dataset = [[3.999208922,1.209014212,0],
           [8.497545867,2.162953546,1],
           [10.00220326,2.339047188,1],
           [2.728571309,0.169761413,0],
           [4.678319846,1.81281357,0],
           [11.12493903,2.234550982,1],
           [3.771244718,0.784783929,0],
           [4.961043357,1.61995032,0],
           [7.642287351,2.319983761,1],
           [8.444542326,1.476683375,1]]
```

```
classes = list(set(row[-1] for row in dataset))
f_index, f_value, f_score, f_groups = 1, 1, 1, None
for index in range(len(dataset[0])-1):
  for r in dataset:
    left, right = [], []	
    for row in dataset:
      if row[index] < r[index]:
        left.append(row)
      else:
        right.append(row)
      groups = [left,right]


      ni = float(sum([len(group) for group in groups]))

      gini = 0.0
      for group in groups:       
        size = float(len(group))
        if (size == 0):
          continue
        score = 0.0
        for class_val in classes:
          p = [row[-1] for row in group].count(class_val) / size
          score += p * p
        gini += (1.0 - score) * (size / ni)
    print('X%d < %.3f Gini=%.3f' % ((index+1), r[index], gini))

    if gini < f_score:
      f_index, f_value, f_score, f_groups = index, r[index], gini, groups
print('index',f_index, 'value',f_value, 'groups',f_groups)
```

```
k = [7.8319846, 2.81281357]

if k[f_index] < f_value:
  g = 0
  pred = [row[-1] for row in f_groups[g]]
  print(max(pred , key=pred.count))
else:
  g = 1
  pred = [row[-1] for row in f_groups[g]]
  print(max(pred , key=pred.count))
```
